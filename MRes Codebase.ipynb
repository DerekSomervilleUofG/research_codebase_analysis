```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
def get_integer(number):
    result = 0
    if number.isnumeric():
        result = int(number)
    return result
```

```python
def calc_percentage(number, base):
    percentage = 0
    if get_integer(base) > 0:
        percentage = int(number)/int(base)
    return percentage
```

```python
developer_col = 1
authored_date_col = 2
commit_files_col = 3
known_files_col = 6
```

```python
def calc_percentage_from_row(row):
    return calc_percentage(row[known_files_col], row[commit_files_col])
```

```python
developers = {}
developers_start = {}
with open(path, newline='') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in csv_reader:
        if row[developer_col] in developers:
            if row[authored_date_col].isnumeric():
                developers[row[developer_col]][0].append(int(row[authored_date_col]) - developers_start[row[developer_col]])
            else:
                developers[row[developer_col]][0].append(0)
            developers[row[developer_col]][1].append(row[known_files_col])
            developers[row[developer_col]][2].append(calc_percentage_from_row(row))
        else:
            developers_start[row[developer_col]] = get_integer(row[authored_date_col])
            developers[row[developer_col]] = [[0],[row[known_files_col]],[calc_percentage_from_row(row)]]
print("Loaded", len(developers.keys()))
```
